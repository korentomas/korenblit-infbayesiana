```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('predicciones_atp_con_kelly.csv')
df
```

```python
df = df.loc[df['bankroll'] > 0].copy()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Convert date column to datetime
df['date'] = pd.to_datetime(df['date'])

```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['date'], np.log(df['bankroll']))
plt.title("log(Bankroll) Growth Over Time")
plt.xlabel("Date") 
plt.ylabel("log(Bankroll)")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

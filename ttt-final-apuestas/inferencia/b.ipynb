```python
import pandas as pd
import random
```

```python
df = pd.read_csv("predicciones_atp_pwin.csv")
print(df.columns)
display(df.head())
```

```python
df.rename(columns={"winner": "player_a", "loser": "player_b", "m_winner" : "mu_a", "m_loser" : "mu_b","s_winner": "skill_a", "s_loser": "skill_b", "b_winner" : "b_a", "b_loser" : "b_b", "p_win": "p_a"}, inplace=True)
df['p_b'] = 1 - df['p_a']
df['a_won'] = 1
display(df.head())
```

```python
import numpy as np

# 1. Draw a Boolean mask: True â‰ˆ 50 % of the rows
mask = np.random.rand(len(df)) < 0.5

# 2. Swap the two columns *only* where the mask is True
df.loc[mask, ['player_a', 'player_b']] = (
    df.loc[mask, ['player_b', 'player_a']].to_numpy()
)

df.loc[mask, ['b_a', 'b_b']] = (
    df.loc[mask, ['b_b', 'b_a']].to_numpy()
)

# 3. Update the outcome for those swapped rows
df.loc[mask, 'a_won'] = 0
```

```python
df.to_csv("predicciones_atp_pwin_reordered.csv", index=False)
```
